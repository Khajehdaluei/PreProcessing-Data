```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# data_normal = stats.norm.rvs(loc=0.0, scale=1.0, size=1000)
data_normal = np.random.normal(loc=0.0, scale=1.0, size=1000)
sns.histplot(data_normal , bins='auto',kde=True)
```

```python
# data_uniform = stats.uniform.rvs(loc=5, scale=5, size=1000)
data_uniform = np.random.uniform(low=5, high=10, size=1000)
sns.histplot(data_uniform , bins='auto',kde=True)
```

```python
# data_bernoulli = stats.bernoulli.rvs(loc=0 ,p=0.2 ,size=1000)
data_bernoulli = np.random.binomial(n=1 ,p=0.2 ,size=1000)
sns.histplot(data_bernoulli , bins='auto',kde=True)
```

```python
# data_binom = stats.binom.rvs(n=100, p=0.2 ,size=1000) #n ta moraje e ba ehtemal p
data_binom = np.random.binomial(n=100 ,p=0.2 ,size=1000)
sns.histplot(data_binom , bins='auto', kde=True)
```

```python
icu , surgery , el = 0.3 , 0.2, 0.5
data_multinomial = np.random.multinomial(n=1000, pvals=[ surgery, icu, el], size=365)
sns.histplot(data_multinomial , bins='auto', kde=True)
print(data_multinomial)
```

```python
# data_poison = stats.poisson.rvs(20, size=1000)
data_poison = np.random.poisson(lam=20, size=1000) #int out
sns.histplot(data_poison , bins='auto', kde=True)
```

```python
data_exponential = stats.expon.rvs(20,size=1000)
# data_exponential = np.random.exponential(scale=20, size=1000)
sns.histplot(data_exponential , bins='auto', kde=True)
```
