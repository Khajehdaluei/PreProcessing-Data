```python
from scipy import stats
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../__data/weight-height.csv')
data.tail()
```

###  KS Test
#### Khode Data

```python
H_Mail = data[data['Gender']== 'Male']['Height']
H_Femaie = data[data['Gender']== 'Female']['Height']

print(stats.kstest(H_Mail,H_Femaie))

plt.hist((H_Mail.sort_values()), bins=100)
plt.hist((H_Femaie.sort_values()), bins=100)
plt.figure()
plt.plot((H_Mail.sort_values().reset_index(drop=True)))
plt.plot((H_Femaie.sort_values().reset_index(drop=True)))
plt.figure()
plt.plot(np.cumsum((H_Mail.sort_values().reset_index(drop=True))))
plt.plot(np.cumsum((H_Femaie.sort_values().reset_index(drop=True))))
plt.figure()
```

#### histogram data

```python
m_cont , m_devision = np.histogram(H_Mail,bins=100)
f_count , f_devision = np.histogram(H_Femaie,bins=100)

print(stats.kstest(m_cont,f_count))

plt.plot(m_cont)
plt.plot(f_count)
plt.figure()
plt.plot(np.cumsum(m_cont), color='red')
plt.plot(np.cumsum(f_count), color='blue')
```

#### moghyese ba normal

```python
stats.kstest(H_Mail, 'norm')
```

```python
H_Mail_norm = stats.norm.rvs(size=len(H_Mail), loc=H_Mail.mean(), scale=H_Mail.std())
stats.kstest(H_Mail, H_Mail_norm)
```
