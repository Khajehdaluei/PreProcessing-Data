# Mean Mod Median

## 1- Lib

```python
from sklearn.impute import SimpleImputer 
```

```python
si = SimpleImputer(missing_values=np.nan, strategy='mean') # creat strategy

si.fit(X[:, 1:]) # fiting data

X[:, 1:] = si.transform(X[:, 1:]) # replace fiting

# fit and transform --> fit_transform ex: X[:, 1:] = si.fit_transform(X[:, 1:])
X
```

##2- Manual


```python
skin_mean = data_2[data_2['skin'] != 0]['skin'].mean()
skin_mean
```

```python
data_2.replace({'skin': 0}, skin_mean, inplace=True) # skin_mean or np.nan
data_2
```

# KNN Imputer

```python
from sklearn.impute import KNNImputer
```

```python
imputer = KNNImputer(n_neighbors=3)
X[:, 1:] = imputer.fit_transform(X[:, 1:])
X
```
