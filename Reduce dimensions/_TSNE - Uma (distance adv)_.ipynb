<h1><div style="direction:rtl;text-align:center">Reduce Dimensions</div></h1>

```python
import umap
from sklearn.manifold import TSNE
from sklearn.datasets import make_moons
from matplotlib import pyplot as plt
```

```python
X, y = make_moons(n_samples=1000, noise=0.01)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
```

## TSNE

```python
tsne = TSNE(n_components=1, perplexity=50 ,random_state=123)
tsne_result = tsne.fit_transform(X)
```

```python
plt.scatter(tsne_result[:, 0], range(0, len(tsne_result)), c=y)
```

## umap

```python
import umap
```

```python
u = umap.UMAP(n_components=1)
u_result = u.fit_transform(X)
```

```python
plt.scatter(u_result[:, 0], range(0, len(u_result)), c=y)
```
