<h1><div style="direction:rtl;text-align:center">Open CV</div></h1>

```python
from sklearn.decomposition import PCA, KernelPCA
from sklearn.datasets import make_moons
from matplotlib import pyplot as plt
```

```python
X, y = make_moons(n_samples=1000, noise=0.01)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
```

## PCA

```python
pca = PCA(n_components=1, random_state=123)
pca_result = pca.fit_transform(X)
```

```python
plt.scatter(pca_result[:, 0], range(0, len(pca_result)), c=y)
```

## KernelPCA

```python
kpca = KernelPCA(n_components=1, kernel='poly', gamma=10, random_state=123)
kpca_result = kpca.fit_transform(X)
```

```python
plt.scatter(kpca_result[:, 0], range(0, len(kpca_result)), c=y)
```
